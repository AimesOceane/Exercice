```python
#Import des libraries 

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace( -0.7 , 0.7 , 150 )
y = np.linspace( -0.7 , 0.7 , 150 )
```

```python
a, b = np.meshgrid( x , y )
C = a ** 2 + b ** 2 - 0.2
```

```python
figure, axes = plt.subplots()
axes.contour( a , b , C , [0] )
axes.set_aspect( 1 )
plt.title( 'Center-Radius form Circle' )
plt.show()
```
